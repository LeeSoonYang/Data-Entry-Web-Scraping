```python
# Import packages
import requests
from bs4 import BeautifulSoup

# import pandas library as pd
import pandas as pd
import numpy as np

final = pd.DataFrame()
titles = pd.DataFrame()
locations = pd.DataFrame()
```

```python
source = pd.read_excel('testing file.xlsx')
```

```python
excluded_media = ['']
```

```python
count = -1
for i in range(len(source.index)):
    entry = source.iloc[i]
    url = entry['url']
    remark = entry['source']
    cco = entry['y/n']
    
    if remark in excluded_media:
        d = {'A': ['#'], 'B': ['#'],'C': ['#'],'D': ['#'],'E': ['#'],
            'F': ['#'],'G': ['#'],'H': ['#'],'I': ['#'],
             'J': ['#'],'K': ['#'],'L': ['#'],'M': ['#'],'N': ['#'],
             'Hyperlink': [url],'Remark': [remark]}
        result = pd.DataFrame(data=d)
        if cco=='y':
            result['A']=1
            result['B']=1
            result['C']=1
        final = final.append(result)
        
    else:
        try:
            # Package the request, send the request and catch the response: r
            r = requests.get(url)

            # Extract the response as html: html_doc
            html_doc = r.text

            # Create a BeautifulSoup object from the HTML: soup
            soup = BeautifulSoup(html_doc)

            # Get Guido's text: guido_text
            guido_text = soup.get_text()

            # Get the title of Guido's webpage: guido_title
            guido_title = soup.title.get_text()

            #change title to lower letter
            title = guido_title.lower() 

            # create an Empty DataFrame object
            df = pd.DataFrame()
            df['A'] = '#'

            #change text to lower letter
            text = guido_text.lower()

            #code for headline
            list_ = ['A', 'B','C']
            if any(word in title for word in list_):
                df['A'] = ['1']
            else:
                df['A'] = ['2'] 

            #to retrive paragraph from html
            for tag in soup():
                for attribute in ["A", "A", "A", "A",'A']:
                    del tag[attribute]

            p_tags = soup.find_all('p')
            three_para = p_tags[0:3]

            #transform first 3 paragraph to lower letter
            mystring = ' '.join(map(str,three_para))
            lower_three = mystring.lower()

            #code for school named in first three paragraph
            if any(word in lower_three for word in list_):
                df['A'] = ['1']
            else:
                df['A'] = ['2'] 

            #Fill in empty columns
            df['A'] = '#'
            df['B'] = '#'
            df['C'] = '#'
            df['D'] = '#'

            #code to check for management names
            if 'A' in text and 'B' in text:
                df['A'] = ['6']

            elif 'A' in text and 'C' in text:
                df['A'] = ['7']

            elif 'A' in text and 'D' in text:
                df['A'] = ['8']

            elif 'A' in text:
                df['A'] = ['1']

            elif 'B' in text:
                df['A'] = ['2']

            elif 'C' in text:
                df['A'] = ['3']

            elif 'D' in text:
                df['A'] = ['4']

            else:
                df['A'] = ['5']

            #import excel file to retrieve media values
            import pandas as pd
            file = 'EXCEL FILE.xlsm'
            data = pd.ExcelFile(file)
            df1 = data.parse('A')

            #convert dataframe to series
            df1_1=df1['A'].values.tolist()
            df1_2=df1['A'].values.tolist()
            df1_3=df1['A'].values.tolist()
            df1_4=df1['A'].values.tolist()
            df1_5=df1['A'].values.tolist()

            #code to obtain media label and location
            if remark in df1_1:
                df2=df1.loc[df1['A'] == remark, ['A','B','C']]
            elif remark in df1_2:
                df2=df1.loc[df1['A'] == remark, ['A','B','C']]
            elif remark in df1_3:
                df2=df1.loc[df1['A'] == remark, ['A','B','C']]
            elif remark in df1_4:
                df2=df1.loc[df1['A'] == remark, ['A','B','C']]
            elif remark in df1_5:
                df2=df1.loc[df1['A'] == remark, ['A','B','C']]
            else:
                df2 = pd.DataFrame({'A': ['#'],'B': ['#'],'C': ['#'] })

            #to concate the two dataframes together into one
            df2_list = df2.values.tolist()
            df2_list
            df3 = pd.DataFrame(df2_list, columns = ['A','B','C'])
            result = pd.concat([df, df3], axis=1)

            #check if source is from RSIS
            result['A'] = soup.title.get_text()
            list_2 = ['A', 'B']
            if any(word in guido_text.lower() for word in list_2):
                result['A'] = 11
                result['B'] = 0
                result['C'] = 3
                result['D'] = 2
                result['E'] = 2
                result['F'] = 2
                result['G'] = 5
            else:
                result['A'] = '#'
                result['B'] = '#'

            #adding in url and remark columns
            result['Hyperlink'] = url
            result['Remark'] = remark
            
            #code for CCO initiated
            if cco=='y':
                result['A']=1
                result['B']=1
                result['C']=1 

            t = {'A' : [guido_title]}
            titles = titles.append(pd.DataFrame(data=t))
            locations = locations.append(df2)
            final = final.append(result)
            count += 1
            
        except:
            d = {'A': ['#'], 'B': ['#'],'C': ['#'],'D': ['#'],'E': ['#'],
                'F': ['#'],'G': ['#'],'H': ['#'],'I': ['#'],
                 'J': ['#'],'K': ['#'],
                 'L': ['#'],'M': ['#'],'N': ['#'],'Hyperlink': [url],'Remark': [remark]}
            result = pd.DataFrame(data=d)
            if cco=='y':
                result['A']=1
                result['B']=1
                result['C']=1
            final = final.append(result)
            count += 1
            pass
```
